##Import data into R. Prune to relevant samples (PS, non-bulk). Get rid of OTUs not observed in at least 25% of samples)

```python
%%R 
physeq = import_biom("data/otu_table_wtax.biom", "/home/chantal/Chazy/data/otusn.tree")
#print(physeq)
sample.data = import_qiime_sample_data("data/Mapping_file_bulk_Julysamples.txt")
physeq = merge_phyloseq(physeq,sample.data)
physeq.md = sample_data(physeq)
physeq.PY = prune_samples((physeq.md$Treatment %in% c("P","10Y", "20Y", "C")), physeq)
print(physeq.PY)
physeq.PY.snorm = transform_sample_counts(physeq.PY, function(x) x/sum(x))
physeq.PY.thresh = filter_taxa(physeq.PY.snorm, function(x) sum(x > 0) / length(x) > (0.25), TRUE)
print(physeq.PY.thresh)
```

###Calculate Unifrac distances between samples.

```python
%%R
registerDoParallel(cores=20)
wunif.dist = distance(physeq.PY.thresh, 
                      method = "unifrac", 
                      weighted = TRUE,
                      fast = TRUE, 
                      parallel = TRUE, 
                      normalized = FALSE)
```

###Ordinate using weighted unifrac distnaces with NMDS.

```python
%%R -w 500 -h 350
nmds = ordinate(physeq.PY.thresh, 
                method = "NMDS", 
                distance = wunif.dist)
p.nmds = plot_ordination(physeq.PY.thresh, nmds, justDF = TRUE)
#print(p)

asNumeric = function(x) { as.numeric(as.character(x)) }

p.nmds1 = ggplot(p.nmds, aes(x = NMDS1, y = NMDS2)) +  
        geom_point(aes(fill = Treatment, size = 4, alpha = 0.85), pch=21) +
        guides(alpha = FALSE, fill=guide_legend(override.aes=list(size=6)))

print(p.nmds1)
```

```python
%%R -w 500 -h 350
pcoa = ordinate(physeq.PY.thresh, 
                method = "PCoA", 
                distance = wunif.dist)
p.pcoa = plot_ordination(physeq.PY.thresh, pcoa, justDF = TRUE)
#print(p)

asNumeric = function(x) { as.numeric(as.character(x)) }

p.pcoa1 = ggplot(p.pcoa, aes(x = Axis.1, y = Axis.2)) +  
        geom_point(aes(fill = Treatment, size = 4, alpha = 0.85), pch=21) +
        guides(alpha = FALSE, fill=guide_legend(override.aes=list(size=6)))

print(p.pcoa1)
```

# Checking different distance matrices

```python
%%R
dist_methods = unlist(distance("list"))
dist_methods = dist_methods[-(1:2)]
dist_methods = dist_methods[-which(dist_methods == "ANY")]
dist_methods = dist_methods[-which(dist_methods == "morisita")]
dist_methods = dist_methods[-which(dist_methods == "mountford")]
print(dist_methods)
```

```python
%%R
plist <- vector("list", length(dist_methods))
names(plist) = dist_methods
for (i in dist_methods) {
    print(i) #for debugging
    # Calculate distance matrix
        iDist <- distance(physeq.PY.thresh, method = i)
    # Calculate ordination
    iMDS <- ordinate(physeq.PY.thresh, "MDS", distance = iDist)
    ## Make plot Don't carry over previous plot (if error, p will be blank)
    p <- NULL
    # Create plot, store as temp variable, p
    p <- plot_ordination(physeq.PY.thresh, iMDS, color = "treatment")
    # Add title to each plot
    p <- p + ggtitle(paste("MDS using distance method ", i, sep = ""))
    plist[[i]] = p
}
```

```python
%%R -h 1800 -w 2000
df = ldply(plist, function(x) x$data)
names(df)[1] <- "distance"
p = ggplot(df, aes(Axis.1, Axis.2, color = Treatment))
p = p + geom_point(size = 8, alpha = 0.5)
p = p + facet_wrap(~distance, scales = "free")
p = p + ggtitle("MDS on various distance metrics for RNASIP dataset")
p_facet_dist = p
print(p)
```
