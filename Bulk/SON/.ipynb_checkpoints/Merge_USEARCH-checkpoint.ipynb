```python
%%bash
nprocs=20

tmpdir1=`mktemp -d`
trap "rm -r $tmpdir1" 1 2 3 15
split -d -l 2000000 /var/seq_data/rawseqs_Dec2014/Read1.fastq $tmpdir1/Block

tmpdir2=`mktemp -d`
trap "rm -r $tmpdir2" 1 2 3 15
split -d -l 2000000 /var/seq_data/rawseqs_Dec2014/Read2.fastq $tmpdir2/Block

tmpdir3=`mktemp -d`
trap "rm -r $tmpdir3" 1 2 3 15
ls $tmpdir1 | sed 's:^echo $tmpdir1::'| parallel --gnu -j $nprocs \
-k "usearch -fastq_mergepairs $tmpdir1/{} -reverse $tmpdir2/{} -fastq_minovlen 10 -fastq_minmergelen 50 -fastq_truncqual 2 -fastq_minlen 50 -fastqout $tmpdir3/{}.fastq; cat $tmpdir3/{}.fastq" \
> /var/seq_data/rawseqs_Dec2014/USEARCH_merged_raw-$(date +%F).fastq
rm -r $tmpdir3 $tmpdir2 $tmpdir1
```

```python
%%bash

nprocs=20
maxee=1

tmpdir1=`mktemp -d`
trap "rm -r $tmpdir1" 1 2 3 15
split -d -l 2000000 /var/seq_data/rawseqs_Dec2014/USEARCH_merged_raw-$(date +%F).fastq/Block

tmpdir2=`mktemp -d`
trap "rm -r $tmpdir2" 1 2 3 15
ls $tmpdir1/Block?? | parallel --gnu -j $nprocs -k "usearch -fastq_filter {} \
-fastq_maxee $maxee \
-fastaout $tmpdir2/{#}.fasta >/dev/null 2>&1 && cat $tmpdir2/{#}.fasta" > data/tmp/maxee$maxee.fasta
rm -r $tmpdir2 $tmpdir1

grep -c ">" data/tmp/maxee$maxee.fasta
head -n 8 data/tmp/maxee$maxee.fasta
```
