```python
%%R
library(phyloseq)
library(ggplot2)
library(plyr); library(dplyr); library(tidyr)
library(RColorBrewer)
library(gridExtra)
```

```python
%%R
physeq = import_biom("data/otu_table_SON_wtax.biom", "data/otusn_SON.tree")
sample.data = import_qiime_sample_data("data/Mapping_File_SON.txt")
physeq = merge_phyloseq(physeq,sample.data)
physeq.md = sample_data(physeq)   
physeq.PS.bulk = transform_sample_counts(physeq, function(x) x/sum(x))
#physeq.PS.bulk = filter_taxa(physeq.PS.bulk, function(x) sum(x > 0) > (0.25 * length(x)), TRUE)    
physeq.PS.bulk = filter_taxa(physeq.PS.bulk, function(x) sum(x > 0) > 0, TRUE)    

print(physeq.PS.bulk)
```

```python
%%R
mdf = psmelt(physeq.PS.bulk)
```

```python
%%R
mdf.bymonth = ddply(mdf, .(Month, OTU), summarize, mean_month_abundance = mean(Abundance))
mdf.bymonth = mdf.bymonth[order( mdf.bymonth$Month, -mdf.bymonth$mean_month_abundance),]
mdf.1 = subset(mdf.bymonth, Month == "September")
mdf.1$Rank = rank(-mdf.1$mean_month_abundance, ties.method = "random")
mdf.bymonth.ranked = join(mdf.bymonth, mdf.1[,c("OTU","Rank")], by = "OTU")
```

```python
%%R -w 600 -h 350

p = ggplot(mdf.bymonth.ranked, aes(x = Rank, y = mean_month_abundance, color = factor(Month))) 

p = p + geom_line(alpha = 0.5) 

p = p + geom_line(data = subset(mdf.bymonth.ranked, Month == "September"), size = 2)

p = p + scale_y_log10()

p = p + scale_x_log10()

p = p + theme_bw()

p = p + labs(x = "OTU (September Rank)", y = "Relative Abundance", color = "Month")

p
```

```python
%%R
mdf.bytreat = ddply(mdf, .(Treatment, OTU), summarize, mean_treat_abundance = mean(Abundance))
mdf.bytreat = mdf.bytreat[order( mdf.bytreat$Treatment, -mdf.bytreat$mean_treat_abundance),]
mdf.1 = subset(mdf.bytreat, Treatment == "PTH")
mdf.1$Rank = rank(-mdf.1$mean_treat_abundance, ties.method = "random")
mdf.bytreat.ranked = join(mdf.bytreat, mdf.1[,c("OTU","Rank")], by = "OTU")
```

```python
%%R -w 600 -h 350

p = ggplot(mdf.bytreat.ranked, aes(x = Rank, y = mean_treat_abundance, color = factor(Treatment))) 

p = p + geom_line(alpha = 0.5) 

#p = p + geom_line(data = subset(mdf.bytreat.ranked, Treatment == "PTH"), size = 2)

p = p + scale_y_log10()

p = p + scale_x_log10()

p = p + theme_bw()

p = p + labs(x = "OTU (PTH treatment rank)", y = "Relative Abundance", color = "Treatment")

p
```
