##Import data into R. Prune to relevant samples (PS, non-bulk). Get rid of OTUs not observed in at least 25% of samples)

```python
%%R 
physeq = import_biom("data/otu_table_wtax.biom", "/home/chantal/Chazy/data/otusn.tree")
#print(physeq)
sample.data = import_qiime_sample_data("data/Mapping_file_bulk_Julysamples.txt")
physeq = merge_phyloseq(physeq,sample.data)
physeq.md = sample_data(physeq)
physeq.Chazy = prune_samples((physeq.md$Treatment %in% c("NTH","NTR", "PTH", "PTR")), physeq)
print(physeq.Chazy)
physeq.Chazy.snorm = transform_sample_counts(physeq.Chazy, function(x) x/sum(x))
physeq.Chazy.thresh = filter_taxa(physeq.Chazy.snorm, function(x) sum(x > 0) / length(x) > (0.25), TRUE)
print(physeq.Chazy.thresh)
```

###Calculate Unifrac distances between samples.

```python
%%R
registerDoParallel(cores=20)
wunif.dist = distance(physeq.Chazy.thresh, 
                      method = "unifrac", 
                      weighted = TRUE,
                      fast = TRUE, 
                      parallel = TRUE, 
                      normalized = FALSE)
```

###Ordinate using weighted unifrac distnaces with NMDS.

```python
%%R -w 500 -h 350
nmds = ordinate(physeq.Chazy.thresh, 
                method = "NMDS", 
                distance = wunif.dist)
p.nmds = plot_ordination(physeq.Chazy.thresh, nmds, justDF = TRUE)
#print(p)

asNumeric = function(x) { as.numeric(as.character(x)) }

p.nmds1 = ggplot(p.nmds, aes(x = NMDS1, y = NMDS2)) +  
        geom_point(aes(fill = Treatment), pch=21, size = 4) +
        guides(alpha = FALSE, fill=guide_legend(override.aes=list(size=6)))

print(p.nmds1)
```

```python
%%R -w 500 -h 350
pcoa = ordinate(physeq.Chazy.thresh, 
                method = "PCoA", 
                distance = wunif.dist)
p.pcoa = plot_ordination(physeq.Chazy.thresh, pcoa, justDF = TRUE)
#print(p)

asNumeric = function(x) { as.numeric(as.character(x)) }

p.pcoa1 = ggplot(p.pcoa, aes(x = Axis.1, y = Axis.2)) +  
        geom_point(aes(fill = Treatment), pch=21, size = 4) +
        guides(alpha = FALSE, fill=guide_legend(override.aes=list(size=6)))

print(p.pcoa1)
```

# Checking different distance matrices

```python
%%R
dist_methods = unlist(distance("list"))
dist_methods = dist_methods[-(1:2)]
dist_methods = dist_methods[-which(dist_methods == "ANY")]
dist_methods = dist_methods[-which(dist_methods == "morisita")]
dist_methods = dist_methods[-which(dist_methods == "mountford")]
dist_methods = c(dist_methods, c=c("wunif","uwunif"))
print(dist_methods)
```

```python
%%R
dlist.nmds = vector("list", length(dist_methods))
dlist.mds = vector("list", length(dist_methods))
names(dlist.nmds) = dist_methods
names(dlist.mds) = dist_methods
t = physeq.Chazy.thresh
for (i in dist_methods) {
    # Calculate distance matrix
    if ( i == "wunif" ) {
            registerDoParallel(cores=30)
            iDist <- UniFrac(t, 
                             weighted = TRUE, 
                             fast = TRUE, 
                             normalized = FALSE, 
                             parallel = TRUE)
        } else if ( i == "uwunif" ) {
            registerDoParallel(cores=30)
            iDist <- UniFrac(t, 
                             weighted = FALSE, 
                             fast = TRUE, 
                             normalized = FALSE, 
                             parallel = TRUE)
        } else 
            iDist <- distance(t, method = i)
    
    # Calculate ordination
    iNMDS <- ordinate(t, "NMDS", distance = iDist)
    iMDS <- ordinate(t, "MDS", distance = iDist)

    df.nmds <- NULL
    df.mds <- NULL
    df.nmds <- plot_ordination(t, iNMDS, justDF = TRUE)
    df.mds <- plot_ordination(t, iMDS, justDF = TRUE)  
    dlist.nmds[[i]] = df.nmds
    dlist.mds[[i]] = df.mds
}
```

```python
%%R -h 1800 -w 2000
df.nmds.all = ldply(dlist.nmds, function(x) x)
names(df.nmds.all)[1] <- "distance"
p = ggplot(df.nmds.all, aes(NMDS1, NMDS2, color = Treatment))
#p = p + scale_area(range = c(1,5))
p = p + geom_point(alpha = 0.5, size = 6)
p = p + facet_wrap(~distance, scales = "free")
p = p + ggtitle("NMDS on various distance metrics for C-SIP dataset")
print(p)
```
